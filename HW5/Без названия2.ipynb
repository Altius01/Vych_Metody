```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1+cos(x) = 0

def make_iter_i(x_0, iter_count):
    errors = np.zeros(iter_count)
    for i in range(iter_count):
        x_0 += (np.cos(x_0)+1)/np.sin(x_0)
        errors[i] = np.abs(1+np.cos(x_0))
    return x_0, errors
```

```python
it_count = 10
foo_lin = [1/i for i in range(1, it_count+1)]
foo_quad = [1/i**2 for i in range(1, it_count+1)]
plt.plot(range(it_count), (make_iter_i(1, it_count))[1], label='error')
plt.plot(range(it_count), foo_lin, label='linear')
plt.plot(range(it_count), foo_quad, label='quad')
plt.legend()

(make_iter_i(1, it_count))[0]
```

```python
# x^2 = 2

def make_iter_ii(x_0, iter_count):
    errors = np.zeros(iter_count)
    for i in range(iter_count):
        x_0 = 0.5*(x_0+2/x_0)
        errors[i] = np.abs(x_0**2 - 2)
    return x_0, errors
```

```python
it_count = 10

plt.plot(range(it_count), (make_iter_ii(1, it_count))[1], label='error')
plt.plot(range(it_count), foo_lin, label='linear')
plt.plot(range(it_count), foo_quad, label='quad')
plt.legend()

make_iter_ii(1, it_count)[0]
```
